```python
import glob
import pandas as pd
import json
import re
import glob
import pprint
import numpy as np
# Some plotting
import seaborn as sns
import matplotlib.pyplot as plt

from IPython.display import display
from IPython.display import HTML

%matplotlib inline
```

```python
def load_reports(prefix, equal_length=True):
    report_files = glob.glob('test_fs_protocols/%s*json' % prefix)
    print "Got %d reports" % len(report_files)
    reports = [json.load(open(r)) for r in report_files]
    minnruns = min(len(r) for r in reports) 
    maxnruns = max(len(r) for r in reports) 
    #reports = [r for r in reports if len(r) == nruns]
    if equal_length:
        reports = [ r[:minnruns] for r in reports]
    #reports = [r for r in reports if len(r) == maxnruns]
    print "# of reports per each FS:", [len(r) for r in reports]
    return reports
```

```python
def reports_to_dataframe(reports):
    # we need to expand those 'times' entries with all relevant entries from the reports
    all_reports = []
    for report in reports:
        for run in report:
            run = run.copy()
            run['fs'] = run['fs'].replace("<class '__main__.EXT4'>", "EXT4") # temp quick dirty fix
            times = run.pop('times')
            for icmd, cmdtimes in enumerate(times):
                cmdtimes = cmdtimes.copy()
                cmdtimes[u'icmd'] = icmd
                cmdtimes.update(run)
                command = cmdtimes.pop('command')
                # we are interested only in the "last one" and without specificly the
                # run's testdirectory
                plain_command = re.sub(u"test[0-9]*", u"testX", command.split(';', 1)[-1])
                # also if it contains some address -- not of interest
                plain_command = re.sub(r"<\S* (.*) at 0x\S*>", r"\1", plain_command)
                cmdtimes['command'] = "%02d: %s" % (cmdtimes['icmd'], plain_command)
                warm, cold = cmdtimes.pop('warm'), cmdtimes.pop('cold')
                if warm  >= 0:
                    cmdwarm = cmdtimes.copy()
                    cmdwarm['time'] = warm
                    cmdwarm['run_type'] = 'warm'
                    all_reports.append(cmdwarm)
                cmdtimes['time'] = cold
                cmdtimes['run_type'] = 'cold'

                all_reports.append(cmdtimes)
    print "Got %d entries total" % len(all_reports)
    df = pd.DataFrame(all_reports)
    return df
```

```python
# Some of the cells below are "draft" analyses, so let's preload some data for them
df = reports_to_dataframe(load_reports('simpleannex-big_ndirs=100_nfiles=1000-'))
# simpleannex_ndirs=20_nfiles=100-
```

# FS Ratings per each operation

One of the interesting aspects could be to look at performances per each command and "rating" of each FSs among the tested collection.  Here in the table you will see the rating (0-is the best) and the actual measure per each command, meaned across runs

```python
def get_ratings(values):
    return [x[1] for x in sorted(zip(np.argsort(values), range(len(values))))]
def get_rating_values(values):
    values_argsort = np.argsort(values)
    return ["%d/%.2f" % (x[1], values[x[0]]) for x in sorted(zip(values_argsort, range(len(values))))]
def rate_filesystems(df, run_type='cold'):
    dfm = df.pivot_table(index=['fs', 'command'], columns=['run_type'], values=['time'],  aggfunc=np.mean)
    dfm = dfm[('time', run_type)]
    dfm.dropna(inplace=True)
    dfm_per_cmd = dfm.unstack(level=0).T
    # select few representatives ones we are interested in
    #dfm_per_cmd = dfm_per_cmd.ix[['%s_LVM_MD_raid6' % fs for fs in ['EXT4', 'BTRFS', 'XFS', 'ReiserFS']]+['ZFS_layout=raid10_ashift=12_compression=on_sync=standard']]#, EXT4_LVM_MD_raid6]]
    dfm_per_cmd_rating_values = dfm_per_cmd.copy()
    dfm_per_cmd_ratings = dfm_per_cmd.copy()

    for icol in xrange(len(dfm_per_cmd.columns)):
        dfm_per_cmd_rating_values[dfm_per_cmd.columns[icol]] = get_rating_values(dfm_per_cmd.icol(icol).values)
        dfm_per_cmd_ratings[dfm_per_cmd.columns[icol]] = get_ratings(dfm_per_cmd.icol(icol).values)
    dfm_per_cmd_rating_values.insert(0, 'rating', dfm_per_cmd_ratings.median(axis=1))
    return dfm_per_cmd_rating_values
```

```python
def summary_plots(df, run_type='cold', gross_violins=False):
    display(HTML("<H3>Gross-rating/timing for each filesystem/per each command</H3>"))
    display(rate_filesystems(df, run_type=run_type))
    
    if gross_violins:
        plt.figure(figsize=(15, 7))
        display(HTML("<H3>Gross time per each file-system</H3>"))
        sns.violinplot(df.time, groupby=[df.fs, df.run_type], vert=False, bw=0.4);
        plt.show();

    df_ = df[df['run_type'] == run_type]
    ckwargs = dict(x="run", y="time", data=df_, sharey=False, sharex=False,
               col_wrap=3, size=7, lowess=True)

    # sns.lmplot(col="fs", hue="command", **ckwargs);  # not that interesting/easy to analyze actually
    display(HTML("<H3>Execution time per each command across file-systems and across runs</H3>"))
    #plt.figure();
    sns.lmplot(col="command", hue="fs", **ckwargs);
    plt.show();
```

## Small repositories test

Initial test consisted of creating relatively small repositories each containing 100 files in 20 directories added to git-annex.  Multiple rounds (without removing previously created test directories/repositories) were ran, thus slowly growing impact on the filesystem. ZFS filesystem setups ran only 10 rounds of such tests, while 100 times for other filesystems.

Following results present timings on the first 10 rounds across file systems on "cold" runs of the commands.  Ratings/timing table presents overall rating (0 -- is the best, estimated as a median among ratings across all benchmark commands) with mean timing (across rounds) per each command.

```python
df = reports_to_dataframe(load_reports('simpleannex_ndirs=20_nfiles=100-', equal_length=True))
summary_plots(df, 'cold', gross_violins=True)
```

and "warm" (re-running the same command thus utilizing possible caching of meta-information and data) runs:

```python
summary_plots(df, 'warm')
```

In both of the above reports (cold and warm) on small repositories we can see that ZFS performs quite poorely (and also becomes slower with higher FS utilization on many commands) but still on the same order as XFS and EXT4, while BTRFS and RaiserFS perform much smoother with RaiserFS being the most efficient.

Following plots shows all 100 rounds:

```python
df = reports_to_dataframe(load_reports('simpleannex_ndirs=20_nfiles=100-', equal_length=False))
summary_plots(df, 'cold')
```
