```python
import pandas as pd
import json
import re
import glob
import pprint
import numpy as np
```

```python
nruns = 100 # so I could filter out reports more
#report_files = glob.glob('test_fs_protocols/simpleannex_ndirs=20_nfiles=100-*json')
report_files = glob.glob('test_fs_protocols/simpleannex-big_ndirs=100_nfiles=1000-*json')
print "Got %d reports" % len(report_files)
```

```python
reports = [json.load(open(r)) for r in report_files]
minnruns = min(len(r) for r in reports) 
maxnruns = max(len(r) for r in reports) 
#reports = [r for r in reports if len(r) == nruns]
reports = [ r[:minnruns] for r in reports]
#reports = [r for r in reports if len(r) == maxnruns]

print "# of reports per each FS:", [len(r) for r in reports]
#print [r[-1]['run'] for r in reports]
```

```python
# we need to expand those 'times' entries with all relevant entries from the reports
all_reports = []
for report in reports:
    for run in report:
        run = run.copy()
        run['fs'] = run['fs'].replace("<class '__main__.EXT4'>", "EXT4") # temp quick dirty fix
        times = run.pop('times')
        for icmd, cmdtimes in enumerate(times):
            cmdtimes = cmdtimes.copy()
            cmdtimes[u'icmd'] = icmd
            cmdtimes.update(run)
            command = cmdtimes.pop('command')
            # we are interested only in the "last one" and without specificly the
            # run's testdirectory
            plain_command = re.sub(u"test[0-9]*", u"testX", command.split(';', 1)[-1])
            # also if it contains some address -- not of interest
            plain_command = re.sub(r"<\S* (.*) at 0x\S*>", r"\1", plain_command)
            cmdtimes['command'] = "%02d: %s" % (cmdtimes['icmd'], plain_command)
            warm, cold = cmdtimes.pop('warm'), cmdtimes.pop('cold')
            if warm  >= 0:
                cmdwarm = cmdtimes.copy()
                cmdwarm['time'] = warm
                cmdwarm['run_type'] = 'warm'
                all_reports.append(cmdwarm)
            cmdtimes['time'] = cold
            cmdtimes['run_type'] = 'cold'
            
            all_reports.append(cmdtimes)
print "Got %d entries total" % len(all_reports)
df = pd.DataFrame(all_reports)
```

```python
# What were overall the fastest FS?
df.groupby(['fs'])['time', 'run_type'].sum().sort('time')
```

So what would be the "overall" fastest FS?  Of cause "overall" here is quite a crude measure since we clamp together completely different operations, some of which were quite lenthy and others quite rapid.

# FS Ratings per each operation

One of the interesting aspects could be to look at performances per each command and "rating" of each FSs among the tested collection.  Here in the table you will see the rating (0-is the best) and the actual measure per each command, meaned across runs

```python
def get_ratings(values):
    return [x[1] for x in sorted(zip(np.argsort(values), range(len(values))))]
def get_rating_values(values):
    values_argsort = np.argsort(values)
    return ["%d/%.2f" % (x[1], values[x[0]]) for x in sorted(zip(values_argsort, range(len(values))))]
```

```python
dfm = df.pivot_table(index=['fs', 'command'], columns=['run_type'], values=['time'],  aggfunc=np.mean)
dfm = dfm[('time', 'cold')]
dfm.dropna(inplace=True)
dfm_per_cmd = dfm.unstack(level=0).T
# select few representatives ones we are interested in
#dfm_per_cmd = dfm_per_cmd.ix[['%s_LVM_MD_raid6' % fs for fs in ['EXT4', 'BTRFS', 'XFS', 'ReiserFS']]+['ZFS_layout=raid10_ashift=12_compression=on_sync=standard']]#, EXT4_LVM_MD_raid6]]
dfm_per_cmd_rating_values = dfm_per_cmd.copy()
dfm_per_cmd_ratings = dfm_per_cmd.copy()

for icol in xrange(len(dfm_per_cmd.columns)):
    dfm_per_cmd_rating_values[dfm_per_cmd.columns[icol]] = get_rating_values(dfm_per_cmd.icol(icol).values)
    dfm_per_cmd_ratings[dfm_per_cmd.columns[icol]] = get_ratings(dfm_per_cmd.icol(icol).values)
dfm_per_cmd_rating_values.insert(0, 'rating', dfm_per_cmd_ratings.median(axis=1))
dfm_per_cmd_rating_values
```
